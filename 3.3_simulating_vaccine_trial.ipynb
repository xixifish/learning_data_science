## Simulating a Randomized Trial for a Vaccine
Scientists essentially use an urn model to select the subjects for the treatment and control.
> The mayor was referring to the efficacy rate of the vaccine, which was reported to be about 66%. In comparison, Moderna and Pfizer both reported efficacy rates of about 95% for their vaccines.

**But the scopes of the three clinical trials are not comparable.**

#### Scope
When we evaluate the scope of data, we consider the who, when, and where of the study.

* Included adults age 18 and over, where roughly 40% had preexisting conditions associated with an increased risk for getting severe COVID-19
* Enrolled in the study from October to November 2020
* Came from eight countries across three continents, including the US and South Africa

```43,738 people were enrolled. These participants were split into two groups at random. Half received the new vaccine, and the other half received a placebo.```

```At the end of 28 days, they found 468 cases of COVID-19, with 117 of these in the treatment group and 351 in the control group.```



````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
np.random.seed(42)
````

````python
simulations_fast = np.random.hypergeometric(ngood=468, nbad=43270,
                                            nsample=21869, size=500000)
````

````python
# Create the histogram
plt.figure(figsize=(6, 4))
plt.hist(simulations_fast, bins=50, color='lightgray', edgecolor='white', linewidth=0.5)

# Formatting
plt.xlabel('Cases in the treatment group', fontsize=12)
plt.ylabel('# of simulations', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xlim(180, 280)
plt.ylim(0, 40000)

# Format y-axis to show values like "30k", "20k", etc.
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f"{int(x/1000)}k" if x >= 1000 else str(int(x))))

plt.tight_layout()
plt.show()
````

**We found that not one of the 500,000 simulations had 117 or fewer cases.**

It would be an extremely rare event to see so few cases of COVID-19, if in fact the vaccine was not effective.
